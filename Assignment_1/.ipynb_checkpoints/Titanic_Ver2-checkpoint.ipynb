## Survival Analysis: Age Contribution

* Since it's probable that children were saved first. Creating a new column with a categorical variable Child
* Child will take the **value 1** in cases where **age is less than 18**, and a **value 0** in cases where **age is greater than or equal to 18**

```python
train["Child"] = float('NaN')
train.head()
```

```python
train.loc[train.Age < 18, ["Child"]] = 1
train.head(10)
```

```python
train.loc[train.Age >=18, ["Child"]] = 0
train.head(10)
```

* Normalized Survival Rates for passengers under 18:

```python
print(train["Survived"][train["Child"] == 1].value_counts(normalize = True))
```

* Normalized Survival Rates for passengers 18 or older:

```python
print(train["Survived"][train["Child"] == 0].value_counts(normalize = True))
```

As you can see from the survival proportions: 
* a Child has a survival rate of **53.98%** as compared to **38.10%** of an Adult. Therefore, age does certainly play a role.
